üåÄ Bubble Sort ‚Äì Sorting Student IDs Like Organizing a Messy Desk
Imagine you have a stack of student records, all jumbled up. Your job? Sort them by student ID‚Äîbut you can only compare two at a time! That‚Äôs Bubble Sort in action.

üîé How It Works
1Ô∏è‚É£ Start at the top of the list. 2Ô∏è‚É£ Compare each student‚Äôs ID with the next one. 3Ô∏è‚É£ If they‚Äôre in the wrong order, swap them! 4Ô∏è‚É£ Keep repeating this until the entire list is neatly sorted.

üìå What Makes Bubble Sort Unique?
‚úÖ Looks inside each student‚Äôs record to check the ID, rather than just sorting raw numbers. ‚úÖ Swaps entire student entries‚Äînot just IDs, but names, departments, everything!

‚ö†Ô∏è The Downsides
üí® Slow for big lists‚ÄîBubble Sort does a lot of unnecessary swaps. üîÅ Even if most IDs are sorted, it still checks every single one, making extra work. üß† Best case? If the list is already sorted, it runs faster with fewer checks.

‚è≥ How Efficient Is It?
Time Complexity: O(n¬≤) (not great for large datasets).

Memory Use: More swaps = more memory operations.

üîπ Bubble Sort is like tidying your desk one item at a time. It works, but if the desk is really messy (aka a big, unsorted list), you'll wish you had a better method!



```python
def bubble_sort_by_id(student_list):
    n = len(student_list)
    for i in range(n):
        for j in range(0, n - i - 1):
            if student_list[j]['id'] > student_list[j + 1]['id']:
                student_list[j], student_list[j + 1] = student_list[j + 1], student_list[j]
    return student_list
```

### **üìö Insertion Sort ‚Äì Sorting Students by Name Like Arranging Books on a Shelf**  

Imagine you're organizing a bookshelf, but instead of numbers, you're sorting books **alphabetically** by their titles. You pick up a book, find the right spot, and insert it‚Äîshifting others as needed. That‚Äôs **Insertion Sort** in action!

#### **üîé How It Works**  
1Ô∏è‚É£ Start with the first student‚Äîit‚Äôs already in the right place.  
2Ô∏è‚É£ Take the next student and **compare their name alphabetically** with the previous ones.  
3Ô∏è‚É£ If their name belongs earlier in the list, **shift other entries** to make room.  
4Ô∏è‚É£ Repeat until every student is neatly arranged.  

#### **üìå Key Observations**  
‚úÖ **Looks inside each student dictionary** to access and compare names.  
‚úÖ **Minimal swaps**‚Äîinstead, it mostly shifts entries around.  
‚úÖ **Stable sorting**‚Äîif two students have the same name, their original order is preserved!  

#### **‚ö†Ô∏è Limitations**  
üí® **Slower for big lists**‚Äîlike sorting a huge library one book at a time.  
üîÅ Works best when the list is **almost sorted**‚Äîotherwise, the shifting slows things down.  

### **‚è≥ Performance Breakdown**
- **Time Complexity**: **O(n¬≤)** (not ideal for huge datasets).  
- **Efficient on Small Data**: Great when working with just a handful of students.  



```python
def insertion_sort_by_name(student_list):
    for i in range(1, len(student_list)):
        key = student_list[i]
        j = i - 1
        while j >= 0 and student_list[j]['name'].lower() > key['name'].lower():
            student_list[j + 1] = student_list[j]
            j -= 1
        student_list[j + 1] = key
    return student_list
```


üìä Sorting by CGPA ‚Äì Like Ranking the Top Performers in a Competition
Think of a lineup of students all competing for the top spot based on their CGPA. Instead of shuffling everything around like Bubble Sort, this method picks the highest CGPA each round and moves that student to the front‚Äîlike announcing a winner in each round!

üîé How It Works
1Ô∏è‚É£ Scan the entire list to find the student with the highest CGPA. 2Ô∏è‚É£ Move that student to the front. 3Ô∏è‚É£ Repeat for the remaining students until everyone is ranked in descending order.

üìå What‚Äôs Happening Under the Hood?
‚úÖ Compares CGPA values inside student records. ‚úÖ Only one swap per round‚Äîway more efficient than Bubble Sort! ‚úÖ Moves entire student entries rather than just their CGPA values.

‚ö†Ô∏è Limitations
üí® Still O(n¬≤) complexity, so it‚Äôs not the fastest method. üîÅ Best when swaps are expensive‚Äîsince it only does one swap per round instead of many.

üöÄ Performance Breakdown
Time Complexity: O(n¬≤) (works best on small datasets).

Sorted from Highest to Lowest CGPA‚Äîno extra memory needed!

```python

def selection_sort_by_cgpa_desc(student_list):
    n = len(student_list)
    for i in range(n):
        max_idx = i
        for j in range(i + 1, n):
            if student_list[j]['cgpa'] > student_list[max_idx]['cgpa']:
                max_idx = j
        student_list[i], student_list[max_idx] = student_list[max_idx], student_list[i]
    return student_list
```

# üîÅ Loop until user chooses to exit


```python

while True:
    print("\nWelcome to the Student Management System")
    print("""
         1. Sort by ID
         2. Sort by Name
         3. Sort by CGPA
         4. Exit
    """)    
    choice = input("Enter your choice (1-4): ")

    if choice == "1":
        sorted_by_id = bubble_sort_by_id(students.copy())
        print("\nStudents sorted by ID (Bubble Sort):")
        for student in sorted_by_id:
            print(student)

    elif choice == "2":
        sorted_by_name = insertion_sort_by_name(students.copy())
        print("\nStudents sorted by Name (Insertion Sort):")
        for student in sorted_by_name:
            print(student)

    elif choice == "3":
        sorted_by_cgpa = selection_sort_by_cgpa_desc(students.copy())
        print("\nStudents sorted by CGPA Descending (Selection Sort):")
        for student in sorted_by_cgpa:
            print(student)

    elif choice == "4":
        print("Exiting the program. Goodbye!")
        break

    else:
        print("Invalid choice. Please try again.")
```

### **Sorting Algorithms Explained 
#### **üü† Bubble Sort ‚Äì The Slow & Steady Approach**
Imagine you're arranging a line of students based on height, but instead of picking the tallest first, you just keep swapping two people at a time until it looks right. That‚Äôs **Bubble Sort**‚Äîbrute force, but easy to follow.

üîπ **Why it's good:**  
- Super **simple** to understand.  
- Works decently on **small** or nearly sorted lists.  

üîª **Why it's bad:**  
- **Slow** for big lists‚Äîkeeps swapping unnecessarily.  
- Even if most of the list is sorted, it **doesn‚Äôt take advantage** of it.  

#### **üîµ Insertion Sort ‚Äì The "Sorting As You Go" Method**
Picture yourself organizing a deck of cards in your hand: you pick a card, find the right place, and insert it. That‚Äôs **Insertion Sort**‚Äîit builds a sorted list one item at a time.

üîπ **Why it's good:**  
- **Great for nearly sorted lists** (it moves fewer elements).  
- Works efficiently for **small datasets**.  
- **No extra memory** required‚Äîit sorts everything in place!  

üîª **Why it struggles:**  
- **Terrible for large, messy datasets**‚Äîtoo slow.  
- Advanced sorting methods like QuickSort and MergeSort make it look outdated.  

#### **üü¢ Selection Sort ‚Äì The "Pick the Best, One by One" Approach**
Think of **Selection Sort** like choosing players for a sports team: you always pick the best player available first, then the second-best, and so on until the team is complete.

üîπ **Why it's good:**  
- **Simple** to understand and implement.  
- Does **fewer swaps** than Bubble Sort.  

üîª **Where it falls short:**  
- Always makes **O(n¬≤) comparisons**‚Äîeven if the list is mostly sorted.  
- Unlike smarter algorithms, it **doesn‚Äôt adapt** to sorted sections.
